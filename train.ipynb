```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

**Import test and train data sets**

```
train = pd.read_csv('aps_failure_training_set.csv')
train.head()
```

```
test=pd.read_csv("aps_failure_test_set.csv")
```

**DATA CLEANING**

*   converting columns to numeric
*   converting column 1(target) class to binary values





```
#converting columns to numeric
train.iloc[:,1:]=train.iloc[:,1:].apply(pd.to_numeric, errors='coerce')
```

```
train.head()
```

```
test.iloc[:,1:]=test.iloc[:,1:].apply(pd.to_numeric, errors='coerce')
```

*Converting target variable into binary*

```
class_dict = {'neg':0, 'pos':1}
train["class"] = train["class"].apply(lambda line: class_dict[line])
train_total=train["class"].value_counts()
```

```
test["class"] = test["class"].apply(lambda line: class_dict[line])
test_total=test["class"].value_counts()
```

Percentage of Target variables

*   Imbalanced target variables will effect our model a lot. We need to see how is more target variables distributed to take correct metrics to evaluate our model




*   Counting percentage of  "zeros"  in each columns (Train)












```
total_zero=(train == 0).sum()
p=(total_zero/len(train)*100).sort_values(ascending=False)
p.head(33)
```

*   Counting percentage of "NaN" in each columns (Train)



```
total_na=train.isna().sum()
#percentage=perc
perc=(total_na/len(train)*100).sort_values(ascending=False)
perc.head(8)
```

Counting percentage of "NaN" in each columns (Test)

```
total_na=test.isna().sum()
#percentage=perc
perc=(total_na/len(test)*100).sort_values(ascending=False)
perc.head(8)
```

Removing above columns with 65% above NaN value columns from the train and test.

```
train_df = train.loc[:, train.isnull().sum() < 0.65*train.shape[0]]
test_df = test.loc[:, test.isnull().sum() < 0.65*test.shape[0]]
```

*Filling columns with NaN values with mean*

```
train_df=train_df.apply(lambda x: x.fillna(x.mean()),axis=0)
test_df=test_df.apply(lambda x:x.fillna(x.mean()),axis=0)
```

```
train_df.head()
```

**Pre Processing data**


1.   spliting target and independent variables in both train and test
2.  feature scaling



```
#train target and indepent variables
x_train=train_df.iloc[:,1:]
y_train=train_df.iloc[:,0]
#test target and independent variables
x_test=test_df.iloc[:,1:]
y_test=test_df.iloc[:,0]
```

```
#feature scaling 
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

**Data Classification**

*   Naive Bayes
*   XGBoost
*   Random Forest




*Naive Bayes*

```
# Training with Naive Bayes
from sklearn.naive_bayes import GaussianNB
classifier=GaussianNB()
classifier.fit(x_train,y_train)
```

```
#predecting on test set
y_pred=classifier.predict(x_test)
#comparing predicted results vs actual test results
np.column_stack((y_pred,y_test))
```

```
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
accuracy=accuracy_score(y_test,y_pred)
f1score=f1_score(y_test,y_pred,1)
print('The accuracy is {0}, and f1 score is {1}.'.format(accuracy,f1score))
```

*XGBoost*

```
#traning with gradientboosting
from xgboost import XGBClassifier
classifier=XGBClassifier()
classifier.fit(x_train,y_train)
```

```
#predecting on test set
y_pred=classifier.predict(x_test)
#comparing predicted results vs actual test results
np.column_stack((y_pred,y_test))
```

```
from sklearn.metrics import accuracy_score
from sklearn.metrics import fbeta_score,classification_report
accuracy=accuracy_score(y_test,y_pred)
f1score=fbeta_score(y_test,y_pred,1)
print('The accuracy is {0}, and f1 score is {1}.'.format(accuracy,f1score))
```

```
#predecting on test set
y_pred=RF.predict(x_test)
#comparing predicted results vs actual test results
np.column_stack((y_pred,y_test))
```

```
from sklearn.metrics import accuracy_score,precision_score,recall_score
from sklearn.metrics import fbeta_score,classification_report
accuracy=accuracy_score(y_test,y_pred)
f1score=fbeta_score(y_test,y_pred,1)

print('The accuracy is {0}, and f1 score is {1}, precision score is {2},recall score{3}'.format(accuracy,f1score,p,r))
print(classification_report(y_test, y_pred))
```

**Implementation: Model Tuning**

> Fine tune the chosen model(XGboost). Use grid search (GridSearchCV) with at least one important parameter tuned with at least 3 different values



```
# Import 'GridSearchCV', 'make_scorer'
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import make_scorer

# Initialize the classifier
classifier = XGBClassifier(random_state=0)

#parameters
parameters={'learning_rate':[0.1,0.05,0.005],
            'max_depth': np.arange(2,10,2),
            'n_estimators':[100,150,200] }

# Make an fbeta_score scoring object using make_scorer()
scorer = make_scorer(fbeta_score, beta=0.5)

# Perform grid search on the classifier using 'scorer' as the scoring method using GridSearchCV()
grid_obj = GridSearchCV(classifier, parameters, scoring=scorer, n_jobs=10, verbose=10)

# Fit the grid search object to the training data and find the optimal parameters using fit()
grid_fit = grid_obj.fit(x_train, y_train)

# Get the estimator
best_clf = grid_fit.best_estimator_

print("Best parameters: ")

# Make predictions using the unoptimized and model
predictions = (classifier.fit(x_train, y_train)).predict(x_test)
best_predictions = best_clf.predict(x_test)

# Report the before-and-afterscores
print("Unoptimized model\n------")
print("Accuracy score on testing data: {:.4f}".format(accuracy_score(y_test, y_pred)))
print("F-score on testing data: {:.4f}".format(fbeta_score(y_test, y_pred, beta = 1)))
print("\nOptimized Model\n------")
print("Final accuracy score on the testing data: {:.4f}".format(accuracy_score(y_test, best_predictions)))
print("Final F-score on the testing data: {:.4f}".format(fbeta_score(y_test, best_predictions, beta = 1 )))

```
